```python
x :np.ndarray = np.array([1,2,3,4])
y :np.ndarray = np.array([10, 20, 25, 30])
fig , ax = plt.subplots(figsize=(6,4))
ax.plot(x , y)
ax.set(title='Simple plot' , xlabel='x Axis' , ylabel='Y Axis')
ax.legend()
ax.grid(True ,alpha=0.3 )
fig.tight_layout()
# plt.savefig('./images/sample-plot.png')
plt.show()
```

```python
x : npt.NDArray[np.float64] = np.linspace(0,10 ,100)
y: npt.NDArray[np.float64] = np.sin(x)

fig , ax = plt.subplots(figsize=(8,4))
ax.plot(x , y , color='blue' , label='Sine wave' , linewidth=2 )
ax.set_title('Simple sine wave' , fontsize=14 , fontweight='bold')
ax.set_xlabel('Angle radians')
ax.set_ylabel('sin(x)')
ax.legend(loc='upper right')
ax.grid(True , linestyle='--' , alpha=0.6)
fig.tight_layout()
plt.show()
```
