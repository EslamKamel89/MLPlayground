```python
x : npt.NDArray[np.float64] = np.linspace(0,10 ,100)
y: npt.NDArray[np.float64] = np.sin(x)

fig , ax = plt.subplots(figsize=(8,4))
ax.plot(x , y , color='blue' , label='Sine wave' , linewidth=2 )
ax.set_title('Simple sine wave' , fontsize=14 , fontweight='bold')
ax.set_xlabel('Angle radians')
ax.set_ylabel('sin(x)')
ax.legend(loc='upper right')
ax.grid(True , linestyle='--' , alpha=0.6)
fig.tight_layout()
plt.show()
```

# Scatter plot and bar plot
#### we will focus on 
- line plot 
- scatter plot 
- bar plot 
- histogram plot
- subplots

```python
# making figures with numpy arrays
x = np.linspace(0,10 , 100)
fig , ax = plt.subplots()
# create line plot
ax.plot(x , x**2 , label='x = x**2')
ax.set(title='Line plot' , xlabel='x' , ylabel='y')
ax.legend()
plt.show()
```

```python
fig , ax = plt.subplots()
# create scatter plots
ax.scatter(x , np.exp(x) , s=15 , alpha=0.7 , label = 'y = e * x')
ax.set(label='Scatter plot' , xlabel = 'x' , ylabel = 'exp(x)')
ax.legend()
plt.show()
```

```python
# another scatter plot 
fig , ax = plt.subplots()
ax.scatter(x , x * np.sin(x))
```

```python
# make a plot from a dictionary
butter_prices = {
    'Almond Butter' : 10 , 'Peanut Butter' : 8 , 'Cashew Butter' : 12
}
fig , ax = plt.subplots()
ax.bar(butter_prices.keys() , butter_prices.values())
ax.set(title='Butter prices' , ylabel='price ($)' , xlabel='Butter Type');
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4), constrained_layout=True)

ax1.plot(x, x**2, label="x")
ax1.set(title="Line", xlabel="x", ylabel="y"); ax1.legend()

ax2.scatter(x, np.sin(x), s=15, alpha=0.7, label="sin(x)")
ax2.set(title="Scatter", xlabel="x", ylabel="sin(x)"); ax2.legend()

plt.show()
```

```python
prices = {
    'Almond Butter' : 10 , 'Peanut Butter' : 8 , 'Cashew Butter' : 12
}
fig, ax = plt.subplots()
cats, vals = list(prices.keys()), list(prices.values())
bars = ax.bar(cats, vals)

for bar in bars:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
            f"{bar.get_height():.0f}", ha="center", va="bottom")

ax.set(title="Prices", ylabel="$")
fig.tight_layout(); 
plt.show()
```
