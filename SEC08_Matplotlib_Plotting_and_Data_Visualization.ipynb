```python
from typing import TypeVar


T = TypeVar('T')

def pr(val : T , title:str|None = None , description: str|None = None) : # type: ignore
    if title != None :
        print(f" ---------------------- {title} ---------------------- ")
        if description != None :
            print(f" <-- {description} -->")
    print('type: ' , type(val))
    display(val)
    # return val
```

```python
x : npt.NDArray[np.float64] = np.linspace(0,10 ,100)
y: npt.NDArray[np.float64] = np.sin(x)

fig , ax = plt.subplots(figsize=(8,4))
ax.plot(x , y , color='blue' , label='Sine wave' , linewidth=2 )
ax.set_title('Simple sine wave' , fontsize=14 , fontweight='bold')
ax.set_xlabel('Angle radians')
ax.set_ylabel('sin(x)')
ax.legend(loc='upper right')
ax.grid(True , linestyle='--' , alpha=0.6)
fig.tight_layout()
plt.show()
```

```python
# making figures with numpy arrays
x = np.linspace(0,10 , 100)
fig , ax = plt.subplots()
# create line plot
ax.plot(x , x**2 , label='x = x**2')
ax.set(title='Line plot' , xlabel='x' , ylabel='y')
ax.legend()
plt.show()
```

```python
fig , ax = plt.subplots()
# create scatter plots
ax.scatter(x , np.exp(x) , s=15 , alpha=0.7 , label = 'y = e * x')
ax.set(label='Scatter plot' , xlabel = 'x' , ylabel = 'exp(x)')
ax.legend()
plt.show()
```

```python
# another scatter plot 
fig , ax = plt.subplots()
ax.scatter(x , x * np.sin(x))
```

```python
# make a plot from a dictionary
butter_prices = {
    'Almond Butter' : 10 , 'Peanut Butter' : 8 , 'Cashew Butter' : 12
}
fig , ax = plt.subplots()
ax.bar(butter_prices.keys() , butter_prices.values())
ax.set(title='Butter prices' , ylabel='price ($)' , xlabel='Butter Type');
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4), constrained_layout=True)

ax1.plot(x, x**2, label="x")
ax1.set(title="Line", xlabel="x", ylabel="y"); ax1.legend()

ax2.scatter(x, np.sin(x), s=15, alpha=0.7, label="sin(x)")
ax2.set(title="Scatter", xlabel="x", ylabel="sin(x)"); ax2.legend()

plt.show()
```

```python
prices = {
    'Almond Butter' : 10 , 'Peanut Butter' : 8 , 'Cashew Butter' : 12
}
fig, ax = plt.subplots()
cats, vals = list(prices.keys()), list(prices.values())
bars = ax.bar(cats, vals)

for bar in bars:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
            f"{bar.get_height():.0f}", ha="center", va="bottom")

ax.set(title="Prices", ylabel="$")
fig.tight_layout(); 
plt.show()
```

# Plotting from pandas Dataframe

```python
ts = pd.Series(np.random.randn(1000) , index = pd.date_range('1/1/2020' , periods=1000))
ts = ts.cumsum()
ax =  ts.plot(figsize=(9,4) , title="Cumulative Random Series (Pandas  Matplotlib)")
ax.set_xlabel('Date')
ax.set_ylabel('value')
plt.show()
```

```python
make_counts:pd.Series = car_sales['Make'].value_counts()
pr(make_counts)
ax = make_counts.plot.bar(
    figsize= (9,4),
    title="Car make count",
    rot=0 ,
    color="#4C72B0",
    edgecolor="black",
)
ax.set_xlabel('Make')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

# Plotting From Pandas DataFrames 2

```python
sales = car_sales.copy()
# remove the `$` , `,` , `.` signs from the price column
sales['Price'] = sales['Price'].str.replace(r'[$,.]' , '' , regex=True)
# remove the two zeros from the right.
sales['Price'] = sales['Price'].str[:-1].astype('int')
# add Sale Date column
sales['Sales Date'] = pd.date_range('1/1/2020' , periods=len(sales))
# let's add a total column
sales['Total'] = sales['Price'].cumsum()
pr(sales)
```

```python
# let's plot the total sales
ax = sales.plot(x='Sales Date' , y='Total')
ax.set_xlabel('Sale Date')
ax.set_ylabel('Total Sales ($)')
plt.tight_layout()
plt.show()
```
